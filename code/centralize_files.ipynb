```python
import os
import shutil

from tkinter import filedialog, Tk

```

<h1>Get original path of data from user</h1>

```python
#Get directory of data
root = Tk()
source_dir = filedialog.askdirectory(title="----------------------------------------------------  Select location of 'archive' folder  ----------------------------------------------------")

if(source_dir):
    print(f"Selected directory: {source_dir}")
else:
    print("You must select a directory")

root.destroy()
```

<h1>Move all the data into a centralized folder</h1>

```python
data_folders = ["test", "train", "val"]
sub_folders = ["NORMAL", "PNEUMONIA"]

try:
    for folder_name in data_folders:
        for sub_folder_name in sub_folders:
            directory = os.path.join(source_dir, "chest_xray", folder_name, sub_folder_name)
            directory = os.path.normpath(directory) #Fix mixed up path ('/' and '\' mixing)
            
            for file_name in os.listdir(directory): 
                    source_file = os.path.join(directory, file_name)

                    #Organize file based on label
                    destination_dir = os.path.join(os.path.dirname(source_dir), "data")
                    if "virus" in file_name:
                        destination_dir = os.path.join(destination_dir, "viral")
                    elif "bacteria" in file_name:
                        destination_dir = os.path.join(destination_dir, "bacterial")
                    else:
                        destination_dir = os.path.join(destination_dir, "normal")

                    #Create path if it doesn't already exist
                    os.makedirs(destination_dir, exist_ok=True)

                    # Move the file
                    shutil.move(source_file, destination_dir)
    
    shutil.rmtree(source_dir) #Remove archive folder
except FileNotFoundError:
    print(f"Error: Source file '{source_file}' not found.")
except OSError:
     print(f"Deletion of {source_dir} failed")
```
